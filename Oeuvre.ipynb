```python
import pygame
from PIL import Image
import datetime
import math

# Initialiser Pygame
pygame.init()

# Paramètres de la fenêtre
LARGEUR, HAUTEUR = 600, 600
ecran = pygame.display.set_mode((LARGEUR, HAUTEUR))
pygame.display.set_caption("Horloge avec fond en pixel art")

# Charger et convertir une image en pixel art
def pixel_art(chemin_image, taille):
    img = Image.open(chemin_image)
    # Redimensionner l'image
    img = img.resize(taille, Image.NEAREST)
    img = img.resize((LARGEUR, HAUTEUR), Image.NEAREST)
    # Convertir en format
    return pygame.image.fromstring(img.tobytes(), img.size, img.mode)

# Charger les images de fond
image_jour = pixel_art("soleil.png", (40, 40))  # pour l'effet pixelisé
image_nuit = pixel_art("lune.jpg", (40, 40))


FPS = 24
horloge = pygame.time.Clock()
police = pygame.font.Font(None, 40)


def calculer_position_aiguille(centre, longueur, angle_deg):
    angle_rad = math.radians(angle_deg - 90)  # Ajuster pour démarrer à 12h
    x = centre[0] + longueur * math.cos(angle_rad)
    y = centre[1] + longueur * math.sin(angle_rad)
    return int(x), int(y)



# Boucle principale
en_cours = True
while en_cours:
    for evenement in pygame.event.get():
        if evenement.type == pygame.QUIT:
            en_cours = False

    # Heure actuelle
    maintenant = datetime.datetime.now()
    heure = maintenant.hour % 12
    minute = maintenant.minute
    seconde = maintenant.second

    # Basculer jour nuit
    if 6 <= maintenant.hour < 18:
        fond = image_jour
        couleur_texte = (0, 0, 0)  # Noir pour la journée
    else:
        fond = image_nuit
        couleur_texte = (255, 255, 255)  # Blanc pour la nuit

    # Image de fond
    ecran.blit(fond, (0, 0))

    # Centre et taille du cadran
    centre = (LARGEUR // 2, HAUTEUR // 2)
    rayon = 200
    pygame.draw.circle(ecran, (255, 255, 255), centre, rayon, 2)

    # Dessiner les chiffres autour du cadran
    for i in range(12):
        # Calculer l'angle pour chaque heure
        angle_affichage = 32 + i * 30
        x, y = calculer_position_aiguille(centre, rayon - 30, angle_affichage)
        texte = police.render(str(i + 1), True, (0, 0, 0))  # Chiffre en noir
        rect_texte = texte.get_rect(center=(x, y))
        ecran.blit(texte, rect_texte)

    # Aiguilles
    couleur_seconde = (255, 100, 100)  # Rouge
    couleur_minute = (100, 100, 255)  # Bleu
    couleur_heure = (200, 200, 200)   # Gris clair

    # Calcul des angles
    angle_seconde = seconde * 6
    angle_minute = minute * 6 + seconde * 0.1
    angle_heure = heure * 30 + minute * 0.5

    # Dessiner les aiguilles
    aiguille_seconde = calculer_position_aiguille(centre, rayon - 20, angle_seconde)
    pygame.draw.line(ecran, couleur_seconde, centre, aiguille_seconde, 2)

    aiguille_minute = calculer_position_aiguille(centre, rayon - 40, angle_minute)
    pygame.draw.line(ecran, couleur_minute, centre, aiguille_minute, 4)

    aiguille_heure = calculer_position_aiguille(centre, rayon - 80, angle_heure)
    pygame.draw.line(ecran, couleur_heure, centre, aiguille_heure, 6)

    # Affichage de l'heure en format numérique
    heure_formatee = f"{maintenant.hour:02d}:{maintenant.minute:02d}:{maintenant.second:02d}"
    texte_heure = police.render(heure_formatee, True, couleur_texte)
    rect_heure = texte_heure.get_rect(center=(LARGEUR // 2, HAUTEUR - 40))
    ecran.blit(texte_heure, rect_heure)

    # Mettre à jour l'écran
    pygame.display.flip()

    # Limiter les FPS
    horloge.tick(FPS)

pygame.quit()
```
